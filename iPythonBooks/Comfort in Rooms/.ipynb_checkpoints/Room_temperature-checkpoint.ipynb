```python
print("\t Connected to Mongo DB")

season = ['winter', 'summer', 't_period_1', 't_period_2']
connection = MongoClient('localhost') #"192.168.6.132"
collection_series = 'filtered_time_series'
collection_metadata = 'metadata'
collection_comfort_room_sax = 'comfort_room_sax'

MONGODB_HOST = 'localhost'  # '192.168.6.132'
MONGODB_PORT = 27017
MONGODB_DB = 'project_db'
connection = MongoClient(MONGODB_HOST, MONGODB_PORT)
collection_series = connection[MONGODB_DB][collection_series]
collection_metadata = connection[MONGODB_DB][collection_metadata]
collection_comfort_room_sax = connection[MONGODB_DB][collection_comfort_room_sax]

projection = {'timestamp': True}
time_query = {}  # all the timeline
tag_query = {}  # all variables
period = '1D'  # '1M'
```

```python
print('Get tag names of the temperature in each room')
tag_query = {'category': 'Temperature', 'floor': {"$in": [1, 2, 3]}}
tag_list = rs.get_tag_names(collection_metadata, tag_query)
tag_outdoor_temp = 'V032_outdoor_temp'
```

```python
print('Get outdoor temperature: {0} . Applying a sliding windows of 48h'.format(tag_outdoor_temp))
df_outdoor_temp = rs.get_tag_values(collection_series, time_query, tag_outdoor_temp, series_format='DF_t')
df_outdoor_temp[tag_outdoor_temp] = pd.to_numeric(df_outdoor_temp[tag_outdoor_temp], errors='coerce')
df_outdoor_temp = df_outdoor_temp.fillna(df_outdoor_temp.interpolate())
#print(df_outdoor_temp[df_outdoor_temp[tag_outdoor_temp]<-2])
df_outdoor_temp = df_outdoor_temp.rolling(window=48, min_periods=48, center=True).mean()
#print(df_outdoor_temp[df_outdoor_temp[tag_outdoor_temp]<-2])
```

```python
print('Get values for temperature for all the rooms (fix nan values)')
df_room_temp = rs.get_tag_values(collection_series, time_query, tag_list, series_format='DF_t')
for tag in tag_list:
    df_room_temp[tag] = pd.to_numeric(df_room_temp[tag], errors='coerce')
    df_room_temp[tag] = df_room_temp.fillna(df_room_temp[tag].mean())
```

```python
h_min = 6
h_max = 20
print('Likely hours for presence of people: from {0} to {1}'.format(h_min, h_max))
df_room_temp['hour'] = df_room_temp.index.hour
mask = (df_room_temp['hour'] >= h_min) & (df_room_temp['hour'] <= h_max)
df_room_temp = df_room_temp[mask]
#print(df_room_temp.head(2))
#print(mask)
```

```python
m1 = 2 / 5.5  # slope 1
m2 = 1.5 / 4.5  # slope 2
m3 = (28-25) / (19-10)

b1 = 26.5 - m1*(17.5)
b2 = 22 - m2*(23.5)
b3 = 28 - m3*(19)

f1 = np.poly1d(np.array([m1, b1]))
f2 = np.poly1d(np.array([m2, b2]))
f3 = np.poly1d(np.array([m3, b3]))

x_sup = list(np.linspace(-4,12,3)) + list(np.linspace(12,17.5,3)) + list(np.linspace(17.5,32,3))  
y_sup = [24.5]*3 + list(f1(np.linspace(12,17.5,3))) + [26.5]*3

x_inf = list(np.linspace(-4,19,3)) + list(np.linspace(19,23.5,3)) + list(np.linspace(23.5,32,3))  
y_inf = [20.5]*3 + list(f2(np.linspace(19,23.5,3))) + [22]*3

x_nat = list(np.linspace(-4,10,3)) + list(np.linspace(10,32,3))
y_nat = [25]*3 + list(f3(np.linspace(10,32,3)))
y_sup = np.array(y_sup) 
y_inf = np.array(y_inf)
y_nat = np.array(y_nat)
```

```python
tag = tag_list[2]
alias = rs.get_alias(collection_metadata, tag)

comfort_list = list()

for idx in df_room_temp.index:

    comfort = rs.get_comfort_bin(df_room_temp[tag].loc[idx],
                                 df_outdoor_temp[tag_outdoor_temp].loc[idx])
    comfort_list.append(comfort)

df_room_temp['comfort'] = comfort_list
mask_c = df_room_temp[df_room_temp['comfort'] == 1].index
mask_u = df_room_temp[df_room_temp['comfort'] == 0].index

y_comf = df_room_temp[tag].loc[mask_c]
x_comf = df_outdoor_temp[tag_outdoor_temp].loc[mask_c]

y_ucomf = df_room_temp[tag].loc[mask_u]
x_ucomf = df_outdoor_temp[tag_outdoor_temp].loc[mask_u]
```

```python
n_samples = 5000
idx = np.random.randint(0,len(df_room_temp.index),n_samples)

trace0 = go.Scatter(
    x=x_sup,
    y=y_sup,
    fill= None,
    mode='lines',
    name = 'Comfortable Area',
    line=dict(
        color='green',
    )
)
trace1 = go.Scatter(
    x=x_inf,
    y=y_inf,
    fill='tonexty',
    mode='lines',
    name = 'Upper limit',
    line=dict(
        color='green',
    )
)
trace2 = go.Scatter(
    x=x_comf,
    y=y_comf,
    mode='markers',
    name = 'Inside comfort zone',
    opacity= 0.7,
    line=dict(
        color='green',
    )
)

trace3 = go.Scatter(
    x=x_ucomf,
    y=y_ucomf,
    mode='markers',
    name = 'Outside comfort zone',
    opacity= 0.7,
    line=dict(
        color='orange',
    )
)

trace4 = go.Scatter(
    x=x_nat,
    y=y_nat,
    mode='lines',
    name = "Maximum room temperature <br>with natural ventilation",
    line=dict(
        color='red',
    )
)

layout = go.Layout(
    title='Thermal Comfort - SIA 180 [' + alias + ']',
    xaxis=dict(
        title='Moving average outdoor temperature over 48 hours window [°C]',
        dtick= 1,
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        )
    ),
    yaxis=dict(
        title='Perceived Temperature [°C]',
        #range=[400,850],
        dtick= 1,
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        )
    )
)



data = [trace3, trace2, trace0, trace1, trace4]
fig = go.Figure(data=data, layout=layout)
py.iplot(fig)
```
