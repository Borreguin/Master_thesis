```python
from pymongo import MongoClient
import sys
sys.path.append('../../lib')
import rs_common_framework_v4 as rs
import pandas as pd

from __future__ import print_function

import datetime
import collections

import numpy as np
from matplotlib import cm, pyplot as plt
from matplotlib.dates import YearLocator, MonthLocator
import plotly.offline as py
import plotly.graph_objs as go  #important library for the plotly
#import cufflinks as cf
import numpy as np
import math
import pylab as pl

py.init_notebook_mode(connected=False) # run at the start of every ipython notebook to use plotly.offline
from plotly import tools #to do subplots

from hmmlearn.hmm import GaussianHMM
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
warnings.filterwarnings("ignore", category=RuntimeWarning)

print(__doc__)
```

```python
connection = MongoClient('localhost') #"192.168.6.132"
collection_metadata = connection["project_db"]["metadata"] 
collection_series = connection["project_db"]["filtered_time_series"]
collection_daily_vector = connection["project_db"]["vector_daily"]
collection_statistics = connection["project_db"]["statistics_daily"]
```

```python
time_query = dict()
projection_time = dict()
time_df = rs.get_tag_values(collection_series, time_query, projection_time, series_format='DF_t')
start_time = time_df.index[0]._date_repr
end_time = time_df.index[-1]
time_line = pd.date_range(start_time, end_time, freq='1D')
print(start_time, end_time)
```

```python
selected_category = ['A', 'B']
l = []
for timestamp in time_line:

    day_type = ['weekend', 'working_day']
    
    projection = {'R_vector': True}
    filter_query = {"breakout_group" : selected_category[1], 
                    'timestamp': timestamp._date_repr,
                   'day_type': { '$in': day_type } }
    dictionaries = collection_daily_vector.find_one(filter_query, projection)
    if not dictionaries is None:
        l.append(dictionaries['R_vector'])

X = np.array(l)
print(X.shape)
X
```

```python
print("fitting to HMM and decoding ...", end="")
print()
n_componentes = range(2,8)
model_list = []
hidden_states_list = []
for n in n_componentes:
    # Make an HMM instance and execute fit
    model = GaussianHMM(n_components=n, covariance_type="diag", n_iter=1000).fit(X)
    model_list.append(model)
    
    # Predict the optimal sequence of internal hidden state
    h = model.predict(X)
    hidden_states_list.append(h)
    print("done: " + str(n) + " components")
    
```

```python
feature = 'mean'
tag = 'V022_vent02_CO2'
query_tag = {'tagname': tag, 'day_type': { '$in': day_type }}
projection = {feature:True}
df = rs.get_tag_values(collection_statistics,query_tag,projection,series_format='DF_t')
df.info()
```

```python
def angle_convert(angle):
    if angle > 360:
        return angle - 360
    if angle < 0:
        return 360 + angle
    else:
        return angle

class Radar(object):
    
    def __init__(self, fig, titles, labels, id_label =1, rect=None):
        if rect is None:
            rect = [0.05, 0.05, 0.95, 0.95]
        
        self.max_value = 0.5
        self.n = len(titles)
        angles = np.arange(90, 90-360, -360.0/self.n)
        
        angles = [angle_convert(a) for a in angles]
        self.angles = np.array(angles)
        
        self.axes = [fig.add_axes(rect, projection="polar", label="axes%d" % i) 
                         for i in range(self.n)]
        
        #put the title in each axe
        self.ax = self.axes[0]
        self.ax.set_thetagrids(self.angles, labels=titles, fontsize=10)

        for ax in self.axes[1:]:
            ax.patch.set_visible(False)
            ax.grid("off")
            ax.xaxis.set_visible(False)
            ax.yaxis.set_visible(False)

        self.upgrade_max()
        
    def upgrade_max(self):
        for ax, angle, label in zip(self.axes, self.angles, labels):
            n = 5
            values = np.array(range(1,self.max_value, self.max_value/n))
            values = np.round(values,2)
            ax.set_rgrids(values, angle=angle, labels=values)
            ax.spines["polar"].set_visible(False)
            ax.set_ylim(0, self.max_value)
        #ax.set_rgrids(range(1, 8), angle=angle, labels=label)
        #ax.spines["polar"].set_visible(False)
        #ax.set_ylim(0, 8)

    def plot(self, values, *args, **kw):
        
        values = np.r_[values, values[0]]
        
        if self.max_value <  values.max():
            self.max_value = values.max()  
            self.upgrade_max()
        
        angle = np.deg2rad(np.r_[self.angles, self.angles[0]])
        self.ax.plot(angle, values, *args, **kw)
```

```python
week_name = ['Monday', 'Tuesday', 'Wednesday', 'Thursday','Friday', 'Saturday', 'Sunday']

titles = week_name
# This is not needed
labels = [] 

for n in range(0,6):
    
    print('n_comp: ',model_list[n].n_components)
    fig = pl.figure(figsize=(4, 4))
    radar = Radar(fig, titles, labels)
    
    data = []
    for i in range(model_list[n].n_components):
        mask = (hidden_states_list[n] == i)
        day = np.array(df_cmp[mask]['day'])
        
        n_samples = len(day)
        cnt = dict(collections.Counter(day))
        r = [0]*7
        for d, ix in zip(week_name,range(7)):
            if d in cnt.keys():
                r[ix] = round((cnt[d] / n_samples)*100,2)
            
        print(n_samples, r)
        radar.plot(r,  "o-", lw=1, alpha=0.7, label="Cmp" + str(i+1))
        
    radar.ax.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc=3,ncol=3, mode="expand", borderaxespad=1.5)
```
